# Examples on COCO dataset API

## Import common dependencies

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from pycocotools.coco import COCO
```

## Load COCO dataset image

```python
img = mpimg.imread('data/train2017/000000000009.jpg')
plt.imshow(img)

print("img.shape: ", img.shape)
print("img[0][0]: ", img[0][0])
print("type(img[0][0][0]): ", type(img[0][0][0]))
```

## Load COCO dataset labels

```python
dataDir='data/annotations_trainval2017'
dataType='train2017'
annFile='{}/annotations/instances_{}.json'.format(dataDir,dataType)

coco=COCO(annFile)
```

### Load COCO categories

```python
catIds = coco.getCatIds()
catIds.sort()
print("len(catIds):", len(catIds))
print("catIds:", catIds, "\n")

cats = coco.loadCats(catIds)
print("len(cats): ", len(cats))
print("cats[0].keys():", cats[0].keys())
print("cats[0]:", cats[0], "\n")

# show categories
nms = set([cat['supercategory'] for cat in cats])
print('COCO supercategories: \n{}\n'.format(nms))

nms = set([cat['name'] for cat in cats])
print('COCO categories: \n{}\n'.format(nms))

nms = set([cat['id'] for cat in cats])
print('COCO category ids: \n{}\n'.format(nms))
```

### Load COCO annotations

```python
annIds = coco.getAnnIds()
annIds.sort()
print("len(annIds): ", len(annIds))
# print("annIds: ", annIds)
print("annIds[0]: ", annIds[0], "\n")

anns = coco.loadAnns(annIds)
print("len(anns): ", len(anns))
print("anns[0].keys(): ", anns[0].keys())
print("anns[0]: ", anns[0], "\n")

# show mask
mask = coco.annToMask(anns[0]).T
print("mask.shape: ", mask.shape)
plt.imshow(mask)
```

### Load COCO images

```python
imgIds = coco.getImgIds()
imgIds.sort()
print("len(imgIds): ", len(imgIds))
print("imgIds: ", imgIds, "\n")

imgs = coco.loadImgs(imgIds)
print("len(imgs): ", len(imgs))
print("imgs[0].keys(): ", imgs[0].keys())
print("imgs[0]: ", imgs[0])

# show image
img = mpimg.imread('data/train2017/{}'.format(imgs[0]['file_name']))
plt.imshow(img)
```

### Load a mask/image pair

```python
# get first five annotations
annIds = coco.getAnnIds()
annIds.sort()
anns = coco.loadAnns(annIds)
anns = anns[:5]
print("len(anns):", len(anns))
print("anns:", anns)

# get image data
imgIds = []
for ann in anns:
    imgIds.append(ann['image_id'])
imgs = coco.loadImgs(imgIds)
print("len(imgs):", len(imgs))
print("imgs:", imgs)

# create figure
fig = plt.figure(figsize=(10,16))

# plot images
for i, img in enumerate(imgs):
    plt.subplot(5,2,2*i+1)
    _img = mpimg.imread('data/train2017/{}'.format(img['file_name']))
    plt.imshow(_img)

# plot masks
for i, ann in enumerate(anns):
    plt.subplot(5,2,2*i+2)
    mask = coco.annToMask(ann)
    plt.imshow(mask)

plt.show()

```
